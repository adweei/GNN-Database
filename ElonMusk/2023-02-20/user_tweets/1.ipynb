```python
col_name = ['id', 'referenced', 'tweet_id', 'created_at', 'time_delta']
col = []
for i in range(15):
    with open('%d.txt' %i, 'r') as fp:
        lines = fp.readlines()
        for line in lines:
            line = line.replace('\n', '')
            if '=' not in line:
                row = []
                line = line.split(' ')

                if line[1] == 'None':
                    continue
                else:
                    id = str(line[0])
                    ref = str(line[1])
                    tweet_id = str(line[2])
                    time = line[3] + ' ' + line[4]
                    row = [id, ref, tweet_id, time]

                    ref_time = dataG.loc[dataG['id']== int(ref), 'created_at'].values[0]
                    ref_time = datetime.strptime(ref_time, '%a %b %d %H:%M:%S %z %Y')
                    time = datetime.fromisoformat(time)
                    delta = time - ref_time
                    delta = int(delta.total_seconds()/3600)
                    
                    row.append(delta)
                    #print(ref_time, '||', time, '||', delta)
                    col.append(row)

df = pd.DataFrame(col, columns=col_name, dtype='string')
df['referenced'] = df['referenced'].astype('str')
df['tweet_id'] = df['tweet_id'].astype('str') 
df
```

```python
df.to_excel('RetweetTime_Distribution.xlsx', index= False, sheet_name='retweet time')
df = pd.read_excel('RetweetTime_Distribution.xlsx', sheet_name='retweet time')
df
```

```python
from copy import deepcopy
import numpy as np
retweet_time_count  = dict.fromkeys(list(dataG['id'])) #count retweet of each hour period

for k in retweet_time_count.keys():
    df_copy = df.loc[df['referenced'] == k].copy()
    arr = np.zeros(50, dtype=int)

    for index in df_copy.index:
        delta = df_copy.loc[index, 'time_delta']
        if delta <= 48:
            arr[delta] += 1
        else:
            arr[49] += 1
    retweet_time_count[k] = arr

retweet_time_total = deepcopy(retweet_time_count)
retweet_time_ratio = dict.fromkeys(list(dataG['id']))
for k in retweet_time_total:
    arr = retweet_time_total[k]
    for i in range(1, arr.size):
        arr[i] += arr[i-1]
    retweet_time_total[k] = arr

    m = arr.max()
    retweet_time_ratio[k] = np.round(arr/m, 3)

columns = [str(i)+'hr' for i in range(0, 49)]
columns.append('>48hr')
retweet_time_count = pd.DataFrame.from_dict(retweet_time_count, orient='index', columns=columns)
retweet_time_count = retweet_time_count.reset_index()
retweet_time_count = retweet_time_count.rename(columns={'index':'tweet id'})
retweet_time_count['tweet id'] = retweet_time_count['tweet id'].astype('string')


retweet_time_total = pd.DataFrame.from_dict(retweet_time_total, orient='index', columns=columns)
retweet_time_total = retweet_time_total.reset_index()
retweet_time_total = retweet_time_total.rename(columns={'index':'tweet id'})
retweet_time_total = retweet_time_total.sort_values(by= '>48hr', ascending=False)
retweet_time_total['tweet id'] = retweet_time_total['tweet id'].astype('string')

retweet_time_ratio = pd.DataFrame.from_dict(retweet_time_ratio, orient='index', columns=columns)
retweet_time_ratio = retweet_time_ratio.reset_index()
retweet_time_ratio = retweet_time_ratio.rename(columns={'index':'tweet id'})
columns.reverse()
retweet_time_ratio = retweet_time_ratio.sort_values(by= columns, ascending=False)
retweet_time_ratio['tweet id'] = retweet_time_ratio['tweet id'].astype('string')
```
